### Specifying front size 

```python
my_font_size = 15; 
fig = plt.figure(figsize = (5,5/1.5))
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, label = "Parabola", color = 'steelblue', linewidth = 3, 
        marker = 'o', markersize = 5)

# ALWAYS label your axes! 
# customize label location, font size
ax.set_xlabel('$\mu$',loc='right',fontsize = my_font_size)
ax.set_ylabel('$f(\mu)$',loc='top',fontsize = my_font_size)

# display and customize legends
ax.legend(frameon = False,loc = 'best', fontsize = my_font_size, markerscale = 1.2)

# adding a panel label
ax.text(90, 500, "(a)", fontsize = 1.2 * my_font_size)

# turn off spines for two of the four axies 
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(1.5)
    ax.spines[axis].set_position(("axes", -0.02))
for axis in ['top','right']:
    ax.spines[axis].set_visible(False)
    
# We'll also need to turn off the ticks on the axes that we turned off
# and adjust the length and tickness
ax.tick_params(width = 1.5, length = 6)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

# Set axes limits
ax.set_xlim(0, 100)
ax.set_ylim(0, 10000)

plt.tight_layout()
plt.show()
```

### Using pre-defined styles
1. the ggplot style 
2. the hate silver 538 style
3. the bmh style 
4. the xkcd style

Code: 

`with plt.style.context('ggplot'):
    fig = plt.figure(figsize = (5,5/1.6))
    ...`

```python
## the ggplot style (the R style)
my_fontsize = 15
with plt.style.context('ggplot'):
    fig = plt.figure( figsize = (5, 5/1.6) )
    ax = fig.add_subplot(1,1,1)
    
    # We should add a label to our dataset that will go into a legend
    ax.plot(x, y, label = "Parabola")
    
    # Label the axes.
    ax.set_xlabel("x", fontsize = 1.6*my_fontsize)
    ax.set_ylabel("f(x)", fontsize = 1.6*my_fontsize)
    
    # Display legend
    ax.legend(loc='best', frameon=False, fontsize = my_fontsize)
```

```python
# the nate silver style 
with plt.style.context('fivethirtyeight'):
    fig = plt.figure( figsize = (5, 5/1.6) )
    ax = fig.add_subplot(1,1,1)
    
    # We should add a label to our dataset that will go into a legend
    ax.plot(x, y, label = "Parabola")
    
    # Label the axes. 
    ax.set_xlabel("$x$", fontsize = 1.6*my_fontsize)
    ax.set_ylabel("$f(x)$", fontsize = 1.6*my_fontsize)
    
    # Display legend
    ax.legend(loc='best', frameon=False, fontsize = my_fontsize)
```

```python
with plt.style.context('bmh'):
    fig = plt.figure( figsize = (5, 5/1.6) )
    ax = fig.add_subplot(1,1,1)
    
    # We should add a label to our dataset that will go into a legend
    ax.plot(x, y, label = "Parabola")
    
    # Label the axes. 
    ax.set_xlabel("$x$", fontsize = 1.6*my_fontsize)
    ax.set_ylabel("$f(x)$", fontsize = 1.6*my_fontsize)
    
    # Display legend
    ax.legend(loc='best', frameon=False, fontsize = my_fontsize)
```

```python
# the xkcd style
with plt.xkcd():
    fig = plt.figure( figsize = (5, 5/1.6) )
    ax = fig.add_subplot(1,1,1)

    # We should add a label to our dataset that will go into a legend
    ax.plot(x, y, label = "Parabola", linewidth = 3)
    
    # Now we can label the axes. 
    ax.set_xlabel("x", fontsize = 1.6*my_fontsize)
    ax.set_ylabel("f(x)", fontsize = 1.6*my_fontsize)
    
    # Display legend
    ax.legend(loc = 'best', frameon=False, fontsize = my_fontsize)
```

### How to change background color

```python
with plt.style.context('bmh'):
    fig = plt.figure( figsize = (6, 4) )
    ax = fig.add_subplot(1,1,1, facecolor = 'orange')

    # We should add a label to our dataset that will go into a legend
    ax.plot(x, y, label = "Parabola", linewidth = 3)
    
    # Now we can label the axes. 
    # Always label your axes! Who knows what is in the graph otherwise
    ax.set_xlabel("$x$", fontsize = 1.6*my_fontsize)
    ax.set_ylabel("$f(x)$", fontsize = 1.6*my_fontsize)
    
    ax.text(10, 4500, 'JUST KIDDING', fontsize = 40)
    
    # Display legend
    ax.legend(loc = 'best', frameon=False, fontsize = my_fontsize)
```

### Date type: data frame
attributes: columns, index 

Notice that the output of using the `.attribute` approach is a different kind of object: a `Series`.


The other approach is the `key` in `dictionary`:

```python
# load csv 
# panda data type: dataframe 
# attributes: columns, index 
df = pd.read_csv(arts_path)
print(type(df))
#The columns are the labels across the top
print(df.columns) # columns are like the keys in dictionary
print('-------------------------------')
#The indexes run down the side
print( df.index )
print(type(df.columns))
print('-------------------------------')
## two ways to assess data 
# the attribute way
print(type(df.Major))
print()
print(df.Major)
print('-------------------------------')

# the key in dictionary way
print(type(df['Major'])) 
print()
df['Major']
print('-------------------------------')
## accessing particular item in the series 
print( df['Major'][0] )
print()
print(df.Major[1])
```

### Slicing in Pandas 

We can also access slices of rows in a `dataframe` by using the syntax introduced for `lists`

```python
# row slicing 
print(df['Major'][0::2])
print('-------------------------------')
print(df.Major[6::-2])
print('-------------------------------')
print( df[0::2]['Major'] )
```

### `.loc[]`

```python
# column slicing does not work in the conventional way
# use .loc[] or .iloc[] to slice columns (have to specify both rows and cols)
# cannot use it iteratively, 
# e.g., print( df.loc['Major'::4].loc[::2] )
print(type(df.loc[:,'Major':'Total']))
print(df.loc[:,'Major':'Total'])
print('-------------------------------') 
## on series 
print( type(df['Major']) )
print( df['Major'].loc[::2] )
```

### `.iloc[]`
The `.iloc[]` approach is integer based and works on the indices.

If you notice that 'Major' is the third column, then it follows that

```python
print( type(df.iloc[:, 2::4]) )
print()
print( df.iloc[:, 2::4] )
```

### Examples
Select three fields for further study (for example, Arts, Business, and Engineering).

Read the data from the corresponding `.csv` or `.xlsx` file into a `dictionary` of `dataframes` where the `key` is the name of the field.

```python
data_folder = Path.cwd() / 'Data'
college_folder = data_folder / 'College-Majors'
arts_path = college_folder / 'recent_Arts.csv'
bus_path = college_folder / 'recent_Business.csv'
eng_path = college_folder / 'recent_Engineering.csv'
df_arts = pd.read_csv(arts_path)
df_bus = pd.read_csv(bus_path)
df_eng = pd.read_csv(eng_path)

dict_df = {}
dict_df['Arts'] = df_arts
dict_df['Business'] = df_bus
dict_df['Engineering'] = df_eng
```

```python
print(type(dict_df))
dict_df['Arts']
```

Write out a nice table with a count of the number of majors for each field

```python
field_list = ['Arts','Business','Engineering']
major_num_list = []


for field in field_list:
    major_num_list.append(dict_df[field]['Major'].count())
    

print(major_num_list)
```

```python
pd.DataFrame(major_num_list,index=field_list).T
```

```python
major_list = []
for field in field_list: 
    df_field = dict_df[field]
    max_idx = df_field[['Unemployment_rate']].idxmax()[0]
    major_list.append(df_field.Major[max_idx])
print(major_list)
```
